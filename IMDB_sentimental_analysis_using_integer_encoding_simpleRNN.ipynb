```python
# importing the dependencies
from keras.datasets import imdb
from keras.layers import Dense,Flatten,SimpleRNN
from keras import Sequential
```

```python
(x_train,y_train),(x_test,y_test) = imdb.load_data()
```

```python
from keras.utils import pad_sequences
x_train = pad_sequences(x_train,padding='post')
x_test = pad_sequences(x_test,padding='post')
```

```python
len(x_train[0])
```

```python
len(x_train[1])
```

```python
# model
model = Sequential()
model.add(SimpleRNN(32,input_shape=(2494,1),return_sequences=False))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
# compilation
model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
model.fit(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```

```python
# making user model

word_index = imdb.get_word_index()
def sentence_review(text):
  word = text.lower().split()
  encoded = [word_index.get(w,2) for w in word]
  padded = pad_sequences([encoded],maxlen=2494,padding='post',truncating='post')
  return padded
```

```python
review = "The movie was amazing"
prediction = model.predict(sentence_review(review))

score = prediction[0][0]

# Display result
if score >= 0.5:
    print(f"ğŸ™‚ Positive review (Confidence: {score:.2f})")
else:
    print(f"ğŸ™ Negative review (Confidence: {1 - score:.2f})")
```
