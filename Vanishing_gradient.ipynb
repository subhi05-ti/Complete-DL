```python
# importing the dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import keras
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from keras.layers import Dense
from keras.models import Sequential
```

```python
x,y = make_moons(n_samples = 100, noise=0.05,random_state=42)
```

```python
plt.scatter(x[:,0],x[:,1],c=y,s=100)
plt.show()
```

```python
model = keras.Sequential([
    keras.layers.Dense(10,activation = 'sigmoid',input_dim=2),
    keras.layers.Dense(10,activation='sigmoid'),
    keras.layers.Dense(10,activation='sigmoid'),
    keras.layers.Dense(10,activation='sigmoid'),
    keras.layers.Dense(10,activation='sigmoid'),
    keras.layers.Dense(10,activation='sigmoid'),
    keras.layers.Dense(10,activation='sigmoid'),
    keras.layers.Dense(10,activation='sigmoid'),
    keras.layers.Dense(10,activation='sigmoid'),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
old_weight = model.get_weights()[0]
```

```python
# train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.02,random_state=42)
```

```python
model.fit(x_train,y_train,epochs=1)
```

```python
new_weights = model.get_weights()[0]
```

```python
# to know the learning rate value
model.optimizer.get_config()['learning_rate']
```

```python
gradient = (old_weight - new_weights)/0.001
percent_change = abs(old_weight - new_weights)/old_weight
```
