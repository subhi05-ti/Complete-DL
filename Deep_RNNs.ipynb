```python
# importing the dependencies
import tensorflow
from tensorflow import keras
from keras.layers import Embedding,Dense,SimpleRNN
from keras.models import Sequential
from keras.preprocessing.sequence import pad_sequences
from keras.datasets import imdb
```

```python
# loading the dataset
(x_train,y_train),(x_test,y_test) = imdb.load_data(num_words=10000)
```

```python
# padding
x_train_pad = pad_sequences(x_train,maxlen=100)
x_test_pd = pad_sequences(x_test,maxlen=100)
```

# Deep RNN

```python
# model
model = Sequential()
model.add(Embedding(10000,32,input_length=100))
model.add(SimpleRNN(8,return_sequences=True))
model.add(SimpleRNN(8))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.build(input_shape=(None, 100))
model.summary()
```

```python
# model compilation
model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
# model fitting
model.fit(x_train_pad,y_train,epochs=5,validation_data=(x_test_pd,y_test),batch_size=32,validation_split=0.2)
```

# Deep LSTM

```python
# making deep LSTM
from keras.layers import LSTM
model = Sequential()

model.add(Embedding(10000,32,input_length=100))
model.add(LSTM(8,return_sequences=True))
model.add(LSTM(8))
model.add(Dense(1,activation = 'sigmoid'))
```

```python
model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
model.fit(x_train_pad,y_train,epochs=10,validation_data=(x_test_pd,y_test),batch_size=32,validation_split=0.2)
```

# Deep GRU

```python
# making deep gru
from keras.layers import GRU
model = Sequential()

model.add(Embedding(10000,32,input_length=100))
model.add(GRU(8,return_sequences=True))
model.add(GRU(8))
model.add(Dense(1,activation = 'sigmoid'))
```

```python
model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
model.fit(x_train_pad,y_train,epochs=10,validation_data=(x_test_pd,y_test),batch_size=32,validation_split=0.2)
```
