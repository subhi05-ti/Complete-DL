```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/DailyDelhiClimate.csv')
```

```python
df.head()
```

```python
plt.scatter(df['date'],df['meantemp'],color='red')
```

```python
x1 = df['meantemp'].ewm(alpha=0.211).mean()
```

```python
x1
```

```python
df['ewma'] = x1
```

```python
df
```

```python
plt.scatter(df['date'],df['meantemp'],color='red')
plt.plot(df['date'],x1,color='black',label='alpha=0.211')
plt.legend()
plt.show()
```
