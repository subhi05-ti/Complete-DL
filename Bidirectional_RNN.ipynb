```python
# importing the dependencies
import tensorflow as tf
from tensorflow  import keras
from keras.layers import Dense,Flatten,Bidirectional,Embedding, LSTM
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.datasets import imdb
```

```python
(x_train,y_train),(x_test,y_test) = imdb.load_data(num_words=10000)
```

```python
x_train_pad = pad_sequences(x_train,maxlen=100)
x_test_pad = pad_sequences(x_test,maxlen=100)
```

```python
model = Sequential()
model.add(Embedding(10000,128,input_length=100))
model.add(Bidirectional(LSTM(64)))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.build(input_shape=(None, 100))
model.summary()
```

```python
model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)
```

```python
model.fit(x_train_pad,y_train,epochs=10,validation_data = (x_test_pad,y_test))
```
