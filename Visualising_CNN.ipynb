```python
# importing dependencies

from tensorflow.keras.applications.vgg16 import VGG16
```

```python
# loading the model
model = VGG16()
```

```python
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model)
```

```python
for i in range(len(model.layers)):
  # checking for convolutional layer
  if 'conv' not in model.layers[i].name:
    continue

  # getting filter,and bias
  filters,bias = model.layers[i].get_weights()
  print('layer number',i,model.layers[i].name,filters.shape)
```

```python
# retrieving weigths from second hidden layers
filters,bias = model.layers[1].get_weights()
```

```python
# normalizing filter values to 0-1 so we can visualize them
f_min,f_max = filters.min(),filters.max()
filters = (filters - f_min) / (f_max - f_min)
```

```python
# in first layer it has 64 filters but here we are not taking thta much
# filters we are taking 18 filters here
from matplotlib import pyplot
n_filter=6
ix=1
fig = pyplot.figure(figsize=(15,10))
for i in range(n_filter):
  f = filters[:,:,:,i]
  for j in range(3):
    # subplot for 6 filters and 3 channels
    pyplot.subplot(n_filter,3,ix)
    pyplot.imshow(f[:,:,j],cmap='gray')
    ix+=1
# plot filter
pyplot.show()
```

```python
from tensorflow.keras.models import Model
model = Model(inputs = model.inputs, outputs=model.layers[1].output)
```

```python
from tensorflow.keras.utils import load_img, img_to_array
from tensorflow.keras.applications.vgg16 import preprocess_input
from numpy import expand_dims

# Load the image
img = load_img('/content/vk.jpg', target_size=(224, 224))
# Convert the image to array
img = img_to_array(img)
# Expand dimensions to match the expected input shape
img = expand_dims(img, axis=0)
# Preprocess the image for the VGG16 model
img = preprocess_input(img)
```

```python
features = model.predict(img)

fig = pyplot.figure(figsize=(20,15))
for i in range(1,features.shape[3]+1):
  pyplot.subplot(8,8,i)
  pyplot.imshow(features[0,:,:,i-1],cmap='gray')

pyplot.show()
```

```python
model2 = VGG16()
```

```python
layer_index = [2,5,9,13,17]
outputs = [model2.layers[i].output for i in layer_index]

model3 = Model(inputs=model2.input, outputs=outputs)
```

```python
feature_map = model3.predict(img)

for i,fmap in zip(layer_index,feature_map):
  fig = pyplot.figure(figsize=(20,15))
  fig.suptitle('layer_{}'.format(i),fontsize=20)
  for i in range(1,features.shape[3]+1):
    pyplot.subplot(8,8,i)
    pyplot.imshow(fmap[0,:,:,i-1],cmap='gray')

  pyplot.show()
```
